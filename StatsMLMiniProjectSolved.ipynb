```python
df2=pd.read_csv('Golf.csv')
df2.head()
```

H0: The mean of driving distance of current model and previous model is same. ( Ub = Ua) i;e No difference in mean Ub-Ua = 0

H1: there is a significant difference in means of driving distance of current model and previous model is same. ( Ub != Ua) i;e Ub - Ua < 0

```python
# we can do a paired one tail t test

curr= df2['Current']
new= df2['New']

n= len(curr)
x1= np.mean(curr)
std1= np.std(curr)
se1= std1/np.sqrt(n)

x2= np.mean(new)
std2= np.std(new)
se2= std2/np.sqrt(n)

print('current',x1)
print('new',x2)
```

We can see that there is slight decrese in driving distance of the new coated ball
Further testing to see if it is significant enough . 


```python
# Checking normality

stats.shapiro(curr)
```

```python
stats.shapiro(new)
```

Both are normal. Hence t-test can be performed 


```python
T_stats_, pval = stats.ttest_rel(curr, new)
print(T_stats_, pval)
```

```python
# Since we are only concerned about the one half, 

pval/2
```

The p value is = 0.1045
Since P value is greater than 0.05, we fail to reject the null hypothesis.

There is indeed NO SIGNIFICANT difference in driving distance after the coating.

### Recommendation for Par Inc. would be to go ahead with the coating technology as there is no significant difference in driving distance after the coating even after the coating is applied.
